```python
from datasets import load_dataset

# Load the IMDb dataset (train/test split already available)
dataset = load_dataset("imdb")

# Show dataset structure
print(dataset)

# Look at one sample
print(dataset["train"][0])
```

```python
# Take smaller subsets for quick training/testing
train_data = dataset["train"].shuffle(seed=42).select(range(5000))
test_data = dataset["test"].shuffle(seed=42).select(range(2000))

print("Train size:", len(train_data))
print("Test size:", len(test_data))
print("Example:", train_data[0])
```

```python
import re

def clean_text(example):
    text = example["text"].lower()
    text = re.sub(r"<br\s*/?>", " ", text)   # remove HTML line breaks
    text = re.sub(r"[^a-z0-9\s]", "", text)  # remove punctuation
    text = re.sub(r"\s+", " ", text).strip() # normalize spaces
    return {"text": text}

train_data = train_data.map(clean_text)
test_data = test_data.map(clean_text)

print(train_data[0])
```

```python
X_train = [x["text"] for x in train_data]
y_train = [x["label"] for x in train_data]

X_test = [x["text"] for x in test_data]
y_test = [x["label"] for x in test_data]

print(X_train[:2])
print(y_train[:2])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert text to TF-IDF features
vectorizer = TfidfVectorizer(max_features=5000)  # limit to top 5000 words
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

print("Train TF-IDF shape:", X_train_tfidf.shape)
print("Test TF-IDF shape:", X_test_tfidf.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Train the classifier
model = LogisticRegression(max_iter=1000)
model.fit(X_train_tfidf, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Predictions
y_pred = model.predict(X_test_tfidf)

# Accuracy
acc = accuracy_score(y_test, y_pred)
print("Test Accuracy:", acc)

# Detailed report
print(classification_report(y_test, y_pred, target_names=["Negative", "Positive"]))
```

```python
from transformers import AutoTokenizer

# Use DistilBERT pretrained model
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Tokenize datasets
def tokenize_function(example):
    return tokenizer(example["text"], padding="max_length", truncation=True, max_length=256)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",                
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,                # logs every 10 steps
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    report_to="none",                # disables wandb or tensorboard (optional)
    disable_tqdm=False,              # ensure tqdm progress bar shows
)
```

```python
from transformers import AutoModelForSequenceClassification, Trainer
import evaluate
import numpy as np

# 1️⃣ Load the model for binary classification
model_name = "distilbert-base-uncased"
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

# 2️⃣ Load accuracy metric
accuracy = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return accuracy.compute(predictions=predictions, references=labels)

# 3️⃣ Make sure your tokenized datasets are in torch format with "labels"
# Example for a small subset (fast testing)
from datasets import DatasetDict
small_dataset = DatasetDict({
    "train": tokenized_datasets["train"].shuffle(seed=42).select(range(5000)),
    "test": tokenized_datasets["test"].shuffle(seed=42).select(range(2000))
})
small_dataset = small_dataset.rename_column("label", "labels")
small_dataset.set_format("torch")

# 4️⃣ Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_dataset["train"],
    eval_dataset=small_dataset["test"],
    compute_metrics=compute_metrics
)

trainer.train()


# 6️⃣ Evaluate on test set
results = trainer.evaluate()
print("Test Accuracy:", results["eval_accuracy"])
```

```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load fine-tuned model & tokenizer
model_path = "./distilbert-imdb"
model = AutoModelForSequenceClassification.from_pretrained(model_path)
tokenizer = AutoTokenizer.from_pretrained(model_path)

# Define sentiment labels
labels = ["Negative", "Positive"]

# Prediction function
def predict_sentiment(review):
    inputs = tokenizer(review, return_tensors="pt", padding=True, truncation=True, max_length=256)
    with torch.no_grad():
        outputs = model(**inputs)
        probs = torch.nn.functional.softmax(outputs.logits, dim=-1)
    predicted_class = torch.argmax(probs, dim=1).item()
    confidence = probs[0][predicted_class].item()
    return {labels[0]: float(probs[0][0]), labels[1]: float(probs[0][1])}

# Build Gradio UI
demo = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(lines=4, placeholder="Enter a movie review here..."),
    outputs=gr.Label(num_top_classes=2),
    title="🎬 Sentiment Analysis with DistilBERT",
    description="Type a movie review to classify it as Positive or Negative."
)

# Launch app
demo.launch(share=True)
```
